```python
import pandas as pd
import requests
import json

all_users = [
    {'id':1, 'name': 'Naruto', 'news':[]},
    {'id':2, 'name': 'Hinata', 'news':[]},
    {'id':3, 'name': 'Sasuke', 'news':[]},
    {'id':4, 'name': 'Kakashi', 'news':[]},
]
```

```python
df = pd.DataFrame(all_users)

user_ids = df['id'].tolist()
print(user_ids)
```

```python
def get_user(user_id):
    for user in all_users:
        if user['id'] == user_id:
            return user
    return None


users = [user for id in user_ids if (user := get_user(id)) is not None]
print(json.dumps(users, indent=2))
```

```python
from copy import deepcopy

all_users_copy = deepcopy(all_users)

def generate_message(user_id, message):
    for user in all_users_copy:
        if user['id'] == user_id:
            user['news'].append(message)
            return

generate_message(1, 'oi eu sou goku')

print(all_users_copy)
print(all_users)
```

```python
def update_user(original, modified):
    feedback = []

    original_by_id = {u['id']: u for u in original}
    modified_by_id = {u['id']: u for u in modified}

    for user_id, modified_user in modified_by_id.items():
        original_user = original_by_id.get(user_id)

        if not original_user:
            continue

        if original_user['news'] != modified_user['news']:
            original_user['news'] = deepcopy(modified_user['news'])
            feedback.append(
                f"Usuário {original_user['name']} (id={user_id}) foi atualizado"
            )

    return feedback

load_dates = update_user(all_users, all_users_copy)
print(load_dates)
```
